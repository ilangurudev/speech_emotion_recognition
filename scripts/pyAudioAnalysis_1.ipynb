```python
import pandas as pd
import numpy as np
import os
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import audioFeatureExtraction as aF


def getData(file):
    [fs,x]  = audioBasicIO.readAudioFile('wav/'+file)
    x = np.resize(x,(fs*4,))  #instead of zero padding, repeating the same content
    [s,t] = aF.stFeatureExtraction(x,fs,0.050*fs,0.025*fs)
    print(file,np.shape(x),np.shape(s))
    
    df = pd.DataFrame(columns=t)
    for row in range(np.shape(s)[0]):
        df.iloc[:,row] =[s[row,:]]  
    return df
    
df_audio = pd.DataFrame()
audiofiles = os.listdir('wav')
for file in audiofiles:
    df_audio = df_audio.append(getData(file),ignore_index=True)
```

```python
df_audio['FileName'] = audiofiles
df_audio['Type'] = [file.split('_')[0] for file in audiofiles]

x_train = df_audio.loc[df_audio['Type'] == "train"]
x_test = df_audio.loc[df_audio['Type'] == "test"]

x_train = x_train.reset_index()
x_train = x_train.drop('index',axis = 1)

x_test = x_test.reset_index()
x_test = x_test.drop('index',axis = 1)

x_train = x_train.iloc[:,:34]
x_test = x_test.iloc[:,:34]
```

```python
labels = pd.read_csv('ComParE2018_AtypicalAffect.txt', sep="\t")

def func(row):
    return row.split('_')[0]
    
labels['Type'] = labels['file_name'].apply(func)

y_train = labels.loc[labels['Type'] == "train"]
y_test = labels.loc[labels['Type'] == "test"]

y_train = y_train.reset_index()
y_train = y_train.drop('index',axis = 1)

y_test = y_test.reset_index()
y_test = y_test.drop('index',axis = 1)

y_train = y_train['emotion']
y_test = y_test['emotion']
```

```python
df_flat_train = pd.concat([pd.DataFrame(x_train[x].values.tolist()) for x in x_train.columns[:]], axis=1)
df_flat_test = pd.concat([pd.DataFrame(x_test[x].values.tolist()) for x in x_test.columns[:]], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.model_selection import GridSearchCV

def computeSVM(x_train,y_train,x_test,y_test):
    param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001, 0.00001, 10]}
    clf_grid = GridSearchCV(svm.SVC(kernel = 'linear'),param_grid,cv= 5)
    clf_grid.fit(df_flat_train, y_train.values.tolist())
    y_pred = clf_grid.predict(df_flat_test)
    return accuracy_score(y_test, y_pred)

#X_train, X_test, y_train, y_test = train_test_split(df_audio, labels, test_size=0.33, random_state=42)
computeSVM(x_train,y_train,x_test,y_test)
```
